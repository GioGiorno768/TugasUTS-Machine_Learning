```python
import pandas as pd
```

```python
ds = pd.read_csv('ds_diabetes.csv')
```

```python
ds
```

```python
ds.describe()
pd.unique(ds['diabetes'])
```

```python
ds['diabetes'].value_counts()
```

```python
x= ds.drop(columns='diabetes', axis=1)
y= ds['diabetes']

from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

x['gender'] = le.fit_transform(x['gender'])
x['smoking_history'] = le.fit_transform(x['smoking_history'])

x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.2 , random_state=99)
```

```python
from sklearn.tree import DecisionTreeClassifier 
tree= DecisionTreeClassifier(criterion='entropy')
tree.fit(x_train,y_train)
```

```python
y_prediksi = tree.predict(x_test)
y_prediksi
```

```python
from sklearn.metrics import accuracy_score
print("Nilai Akurasi", accuracy_score(y_test,y_prediksi))
```

```python
databaru = pd.read_excel('db_diabetes.xlsx')
newdata = databaru.iloc[:,:]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
newdata['gender'] = le.fit_transform(newdata['gender'])
newdata['smoking_history'] = le.fit_transform(newdata['smoking_history'])

newprediksi = tree.predict(newdata)
print("Hasil prediksi: \n")
for i, prediksi in enumerate(newprediksi,start=1):
    tree = "Diabetes" if prediksi == 1 else "Tidak Diabetes"
    newdata_baris= newdata.iloc[i-1]
    print(f"Data uji ke- {i}: {newdata_baris.tolist()} = \"{tree}\"")
```
