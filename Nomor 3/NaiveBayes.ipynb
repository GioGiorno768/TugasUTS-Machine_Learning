```python
x= ds.drop(columns='output', axis=1)
y= ds['output']
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.2 , random_state=99)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()

model.fit(x_train,y_train)
```

```python
y_prediksi = model.predict(x_test)
y_prediksi
```

```python
import numpy as np
np.array(y_test)
```

```python
from sklearn.metrics import accuracy_score
print("Nilai Akurasi", accuracy_score(y_test,y_prediksi))
```

```python
databaru = pd.read_excel('db_hati.xlsx')
newdata = databaru.iloc[:,:]


newprediksi = model.predict(newdata)
print("Hasil prediksi: \n")
for i, prediksi in enumerate(newprediksi,start=1):
    newdata_baris= newdata.iloc[i-1]
    print(f"Data uji ke- {i}: {newdata_baris.tolist()} = \"{prediksi}\"")
```
